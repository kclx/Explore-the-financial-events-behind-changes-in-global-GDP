```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_statistics=pd.read_csv('../../output/Countries_GDP_prediction_statistics.csv')
```

```python
data_statistics.head()
```

```python
name_x=data_statistics.iloc[:,0]
```

```python
plt.figure(figsize=(24,8))
plt.subplot(131)
plt.title('MAE statistics')
plt.bar(name_x,data_statistics.iloc[:,1],width=0.5)
plt.subplot(132)
plt.title('RMSE statistics')
plt.bar(name_x,data_statistics.iloc[:,2],width=0.5)
plt.subplot(133)
plt.title('R² statistics')
plt.bar(name_x,data_statistics.iloc[:,3],width=0.5)
plt.show()
```
