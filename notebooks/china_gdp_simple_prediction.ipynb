```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

Dataset = pd.read_csv('../data/GDP/Chinese_GDP.csv')
Dataset=Dataset.T
```

```python
# 1978~2022
y=Dataset.iloc[1:,1].values[::-1]
X=np.array([i for i in range(1978,2023)]).reshape(45,1)
```

```python
from sklearn.preprocessing import PolynomialFeatures

Poly_regressor = PolynomialFeatures(degree=2)
Poly_X = Poly_regressor.fit_transform(X)

from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(Poly_X, y)

plt.figure(figsize=(24,8))
plt.scatter(X, y)
plt.plot(X, regressor.predict(Poly_regressor.fit_transform(X)), color='red')
plt.xlabel('Year')
plt.ylabel('GDP of China')
plt.title('GDP')
plt.show()

print(regressor.predict(Poly_regressor.fit_transform([[2023]])))

print(regressor.predict(Poly_regressor.fit_transform([[2035]])))
```
