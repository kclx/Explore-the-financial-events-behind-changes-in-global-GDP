# World Bank records of China's economy compared with data from the Chinese Bureau of Statistics

```python
import numpy as np
import pandas as pd
```

```python
chinese_economy_world_bank=pd.read_csv('../data/GemDataEXTR/CPI_Price_modified.csv').iloc[:,20]
# 1994-2023
chinese_economy_world_bank=np.array(chinese_economy_world_bank)
```

```python
chinese_economy_china=pd.read_csv('../data/GDP/Chinese_household_consumption_level.csv').iloc[6,:]
# 1978-2021
chinese_economy_china=np.array(chinese_economy_china)
```

```python
# 1995～2021
x=[i for i in range(1995,2022)]
x=np.array(x)
```

```python
y_world_bank=(chinese_economy_world_bank[1:28]-chinese_economy_world_bank[:27])/chinese_economy_world_bank[:27]
```

```python
chinese_economy_china=chinese_economy_china[::-1][:44]
```

```python
y_china=(chinese_economy_china[17:44]-chinese_economy_china[16:43])/chinese_economy_china[16:43]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(24,8))
plt.plot(x,y_china,label='chinese_economy_china')
plt.plot(x,y_world_bank,label='chinese_economy_world_bank')
plt.legend(loc='upper left')
plt.show()
```
