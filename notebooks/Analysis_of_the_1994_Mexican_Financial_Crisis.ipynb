# 汇率分析

```python
peso_to_us_dollar_exchange_rate = pd.read_csv('../data/Analysis_of_the_1994_Mexican_Financial_Crisis/Peso to US dollar exchange rate.csv')
```

```python
plt.figure(figsize=(24,8))
# 创建折线图
plt.plot(peso_to_us_dollar_exchange_rate['time'], peso_to_us_dollar_exchange_rate['Peso to US dollar exchange rate'],label='dollar to peso')

# 设置x轴刻度和标签
plt.xticks(rotation=45)  # 标签倾斜显示，避免重叠
plt.xlabel('time')
plt.ylabel('1 dollar to peso exchange rate')
plt.legend(loc='upper left')

# 设置刻度间隔
x_ticks = np.arange(0, len(peso_to_us_dollar_exchange_rate), 12)
x_labels = peso_to_us_dollar_exchange_rate.loc[x_ticks, 'time']

# 设置x轴刻度显示
plt.xticks(x_ticks, x_labels)

# 设置图表标题等其他属性
plt.title('Changes in exchange rates in Mexico')
plt.grid(True)

# 显示图表
plt.show()
```
