```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
china = pd.read_csv('../data/CPI/china_CPI.csv')
us = pd.read_csv('../data/CPI/USA CPI Inflation from 1913-2022.csv')
```

```python
us = np.array(us.iloc[65:-1, -1])
```

```python
china = np.array(china.iloc[:, 1])
```

```python
us = us / us[0]
china = china / china[0]
```

```python
x = np.array([i for i in range(1978, 2022)])
```

```python
plt.figure(figsize=(24, 8))
plt.title('Comparison of CPI changes in China and the United States')
plt.plot(x, us, label='United States', marker='+')
plt.plot(x, china, label='China', marker='*')
plt.legend(loc='upper left')
plt.show()
```

```python
money_supply_china = pd.read_csv('../data/money_supply/money_supply_china.csv')
```

```python
money_supply_china.head()
```

```python
quasi_money_supply_china = np.array(money_supply_china.iloc[12:-1, 5])
```

```python
quasi_money_supply_china = (quasi_money_supply_china[1:] - quasi_money_supply_china[:-1]) / quasi_money_supply_china[
                                                                                            :-1]
```

```python
quasi_money_supply_china_x = np.array([i for i in range(1991, 2022)])
```

```python
fig = plt.figure(figsize=(24, 8))
ax1 = fig.add_subplot(111)
ax1.plot(x, china, label='cpi')
ax1.set_ylabel('cpi', fontdict={'weight': 'normal', 'size': 15})

ax1.set_title("Changes in China's cpi and quasi-currency volume", fontdict={'weight': 'normal', 'size': 15})

ax2 = ax1.twinx()  # this is the important function
ax2.plot(quasi_money_supply_china_x, quasi_money_supply_china, 'r', label='quasi money supply growth rate')
ax2.set_label('quasi money supply growth rate')
ax2.set_ylabel('quasi money supply growth rate', fontdict={'weight': 'normal', 'size': 15})
ax2.set_ylim(0, 1)

fig.legend(loc='upper left')
plt.show()
```

```python
rmb_to_usd_exchange_rate = pd.read_csv('../data/money_supply/RMB to USD exchange rate.csv').iloc[:, 1]
```

```python
rmb_to_usd_exchange_rate_x = np.array([i for i in range(1978, 2023)])
```

```python
fig = plt.figure(figsize=(24, 8))
ax1 = fig.add_subplot(111)
ax1.plot(x, china, label='cpi', c='b')
ax1.set_ylabel('cpi', fontdict={'weight': 'normal', 'size': 15})

ax1.set_title("Analysis of China's RMB/USD Exchange Rate and CPI Growth", fontdict={'weight': 'normal', 'size': 15})

ax2 = ax1.twinx()
ax2.plot(rmb_to_usd_exchange_rate_x, rmb_to_usd_exchange_rate / 100, label='RMB to USD exchange rate', c='c')
ax2.set_ylabel('RMB/USD exchange rate (1 = USD)', fontdict={'weight': 'normal', 'size': 15})

fig.legend(loc='upper left')
plt.show()
```
