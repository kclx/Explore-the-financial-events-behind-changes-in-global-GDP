```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.pyplot import savefig
```

```python
countries_GDP_1996_1997 = pd.read_csv(
    '../data/countries_GDP_world_bank/countries_GDP.csv').iloc[
                          :, [0, 1, 38, 39]]
```

```python
countries_GDP_1996_1997['rate'] = countries_GDP_1996_1997.apply(lambda x: (x['1997'] - x['1996']) / x['1996'], axis=1)
```

```python
countries_GDP_1996_1997[countries_GDP_1996_1997['rate'] > 0]
```

```python
countries_GDP_1996_1997.sort_values("rate", inplace=True)
```
