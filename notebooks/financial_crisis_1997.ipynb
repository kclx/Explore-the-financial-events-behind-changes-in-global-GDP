```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries_GDP_1996_1997=pd.read_csv('/Volumes/KeQing/Documents/营口理工学院/大二下学期/其他/2023年6月/program/data/countries_GDP_world_bank/countries_GDP.csv').iloc[:,[0,1,38,39]]
```

```python
countries_GDP_1996_1997['rate'] = countries_GDP_1996_1997.apply(lambda x: (x['1997'] - x['1996'])/x['1996'], axis=1)
```

```python
countries_GDP_1996_1997[countries_GDP_1996_1997['rate']>0]
```

```python
countries_GDP_1996_1997.sort_values("rate",inplace=True)
```

```python
x_counties=np.array(countries_GDP_1996_1997.iloc[:,0][0:20])
x_counties_simple=np.array(countries_GDP_1996_1997.iloc[:,1][0:20])
y=np.array(countries_GDP_1996_1997.iloc[:,4][0:20])
```
