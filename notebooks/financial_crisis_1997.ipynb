```python
countries_GDP = pd.read_csv(
    '/Volumes/KeQing/Documents/营口理工学院/大二下学期/其他/2023年6月/program/data/countries_GDP_world_bank/countries_GDP.csv')
```

```python
countries_names = countries_GDP.iloc[:, 0]
countries_names = np.array(countries_names)
```

```python
index = 0
for i in countries_names:
    if i in ['Thailand', 'Indonesia', 'Malaysia', 'Korea, Rep.']:
        print(i + '\t' + str(index))
    index += 1
```

```python
countries_GDP_four_1995_1999 = countries_GDP.iloc[[106, 126, 169, 233, ], [0, 37, 38, 39, 40, 41]]
```

```python
x_counties = np.array(countries_GDP_four_1995_1999.iloc[:, 0]).reshape(1, 4)
```

```python
y = np.array(countries_GDP_four_1995_1999.iloc[:, 1:])
```

```python
x = x_counties[0]
y_1995 = y[:, 0].reshape(1, 4)[0]
y_1996 = y[:, 1].reshape(1, 4)[0]
y_1997 = y[:, 2].reshape(1, 4)[0]
y_1998 = y[:, 3].reshape(1, 4)[0]
y_1999 = y[:, 4].reshape(1, 4)[0]
```

```python
plt.figure(figsize=(24, 8))
plt.title('The impact of the 1997 Asian financial crisis')
width = 0.15
po_1995 = [i - 0.75 + 0.5 for i in range(len(x))]
po_1996 = [i - 0.6 + 0.5 for i in range(len(x))]
po_1997 = [i - 0.45 + 0.5 for i in range(len(x))]
po_1998 = [i - 0.3 + 0.5 for i in range(len(x))]
po_1999 = [i - 0.15 + 0.5 for i in range(len(x))]

plt.bar(po_1995, y_1995, width=width, label='1995')
plt.bar(po_1996, y_1996, width=width, label='1996')
plt.bar(po_1997, y_1997, width=width, label='1997')
plt.bar(po_1998, y_1998, width=width, label='1998')
plt.bar(po_1999, y_1999, width=width, label='1999')

plt.xticks(range(len(x)), x)

plt.xlabel('counties')
plt.ylabel('GDP')

plt.legend()



savefig('../output/test01.png',
        dpi='figure',
        format='png',
        # metadata=None,
        # bbox_inches=None,
        # pad_inches=0.1,
        # facecolor='auto',
        # edgecolor='auto',
        # backend=None
        )

plt.show()
```
